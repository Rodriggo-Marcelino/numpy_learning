## 4Ô∏è‚É£ Indexa√ß√£o e Slicing (selecionar dados)

Indexa√ß√£o e slicing servem para **acessar partes espec√≠ficas** de um array NumPy.
Isso √© essencial em **An√°lise de Dados**, porque voc√™ raramente usa o dataset inteiro.

---
```
√≠ndice ‚Üí  0   1   2   3   4
valor  ‚Üí [10, 20, 30, 40, 50]
```

---

- a[0]     = 10
- a[2]     = 30
- a[-1]    = 50 (√∫ltimo elemento)

---

#### Array 1D

````python
a = np.array([10, 20, 30, 40, 50])
````

````python
a[0]
````

````python
a[0:4:2]
````

````python
a[1:4]
````

````python
a[-1]
````

### Filtro por condi√ß√£o (muito importante)

````python
a[a > 25]
````

## 5Ô∏è‚É£ Indexa√ß√£o em matrizes (2D)

````python
m = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
````

````python
m[0, 1] # linha 0, coluna 1
````

````python
m[:, 1] # todas as linhas da coluna 1
````

````python
m[1, :] # linha 1 inteira
````

## 8Ô∏è‚É£ Estat√≠stica por eixo (axis)

````python
m = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
````

````python
np.mean(m, axis=0) # m√©dia das colunas
````

````python
np.mean(m, axis=1) # m√©dia das linhas
````
